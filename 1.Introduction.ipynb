 Matplotlib library is used for making 2D plots from data in arrays.

```python
#Line Plot
from matplotlib import pyplot as plt
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y)
plt.show()
```

```python
#Bar plot
from matplotlib import pyplot as plt
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.bar(x,y)
plt.show()
```

```python
#Histogram
from matplotlib import pyplot as plt
y=[10,5,8,4,2]
plt.hist(y)
plt.show()
```

```python
#Scatter Plot
from matplotlib import pyplot as plt
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.scatter(x,y)
plt.show()
```

```python
#plotting a Scatter plot using the plot() method by providing the marker style as 'o'
from matplotlib import pyplot as plt
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y,'o')
plt.show()
```
