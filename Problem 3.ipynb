```python
import numpy as np
x = np.linspace(-np.pi, np.pi, 15)
C = np.cos(x)
S = np.sin(x)
```

```python
ax = plt.axes([0.1, 0.1, 0.8, 0.8])
ax1 = ax.plot(x,C,'bs:')
ax2 = ax.plot(x,S,'ro-')
ax.legend(labels = ('Cosine Function', 'Sine Function'), loc = 'upper left')
ax.set_title("Trignometric Function")
plt.show()
```
