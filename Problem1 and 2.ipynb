```python
import matplotlib
from matplotlib import pyplot as plt

x = [5, 2, 9, 4, 7]

y = [10, 5, 8, 4, 2]

plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.hist(y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y,'o')
plt.show()
```

```python
labels = ['Cloathing','Food', 'Shelter']
size = [20, 30, 50]
plt.pie(size, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Basic needs expense')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [5, 15, 10, 20, 25]

plt.plot(x,y1, marker = 'o', label='line 1')
plt.plot(x,y2, marker = 'o', label='line 2')
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4, label='Area 1-2')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Chart Example')

plt.legend()
plt.show()
```
